```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('E:/internship/data/test.csv')
data.head(10)
```

```python
print(data.head())
print(data.info())
print(data.describe())
```

```python
data['Age'].fillna(data['Age'].median(), inplace=True)
data['Embarked'].fillna(data['Embarked'].mode()[0], inplace=True)
data['Cabin'].fillna('Unknown', inplace=True)
data['Sex'] = data['Sex'].map({'male': 0, 'female': 1})
data['Embarked'] = data['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
```

```python
sns.countplot(x='Parch', data=data)
plt.title('Survival Count')
plt.show()
```

```python
sns.histplot(data['Age'], kde=True)
plt.title('Age Distribution')
plt.show()
```

```python
sns.countplot(x='Pclass', data=data)
plt.title('Passenger Class Distribution')
plt.show()
```

```python
sns.barplot(x='Sex', y='Parch', data=data)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
sns.barplot(x='Pclass', y='Parch', data=data)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data[data['Parch'] == 0]['Age'], kde=True, color='red', label='Not Survived')
sns.histplot(data[data['Parch'] == 1]['Age'], kde=True, color='blue', label='Survived')
plt.title('Age Distribution by Survival')
plt.legend()
plt.show()
```

```python
sns.catplot(x='Pclass', y='Parch', hue='Sex', kind='bar', data=data)
plt.title('Survival Rate by Passenger Class and Gender')
plt.show()
```
