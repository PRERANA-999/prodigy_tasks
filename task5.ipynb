```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('E:/internship/data/diabetes.csv')
```

```python
print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
sns.pairplot(df, hue='Outcome')
plt.suptitle('Pairplot of Diabetes Dataset', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Diabetes Dataset')
plt.show()
```

```python
df.hist(bins=20, figsize=(20, 15), layout=(4, 3))
plt.suptitle('Distribution of Features in Diabetes Dataset', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(15, 10))
sns.boxplot(data=df)
plt.xticks(rotation=90)
plt.title('Boxplot of Diabetes Dataset')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Glucose', hue='Outcome', data=df, palette='coolwarm')
plt.title('Scatter Plot of Age vs. Glucose')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.violinplot(x='Outcome', y='Glucose', data=df, palette='coolwarm')
plt.title('Violin Plot of Glucose Distribution by Outcome')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='Outcome', data=df, palette='coolwarm')
plt.title('Count Plot of Outcome')
plt.show()
```
